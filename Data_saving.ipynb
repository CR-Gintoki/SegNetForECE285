```python
import os
import shutil

'''
This file extract labeled and original photos in cityscapes dataset. Rename the 
files and put them in the same directory.

The function will save the photos end with 'labelIds.png' in 'gtFine' folder to 
destination path. Also save photos end with 'leftImg8bit.png' in 'leftImg8bit'
folder to the same path. The file name is modified based on the dataset, index
and type of the photos.

des_path: the destination path of the saved data. The directory will be
created automaticly if not exist. Three sub folders: train, test, valid
will be created in this directory.

Please place the folder 'gtFine' and 'leftImg8bit' in the same
folder, and pass the path of 'gtFine' to the function as the argument. 

File transfer walk through the files in 'gtFine' and will 
extract corresponding files which has the same name in 'leftImg8bit'
folder. Save them to the destination path.

05/19/2018

'''


des_path = '/datasets/home/14/214/c6du/data/Cityscape_modified/'
if not os.path.isdir(des_path):
        os.mkdir(des_path)

obj_folder = [des_path + 'train',
              des_path + 'valid',
              des_path + 'test']

for dir_i in obj_folder:
    if not os.path.isdir(dir_i):
        os.mkdir(dir_i)

def file_transfer(dirname):

    result = []#所有的文件
    cnt_train = 0
    cnt_valid = 0
    cnt_test = 0
    
    for maindir, subdir, file_name_list in os.walk(dirname):
        
        path_sep_lst = maindir.split('/')
        
        if path_sep_lst[-2] == 'test':
            path_sep_lst[-3] = 'leftImg8bit'
          
            for filename in file_name_list:
                file_sep_lst = filename.split('_')
                
                if file_sep_lst[-1] == 'labelIds.png':
                    file_sep_lst.pop(-1)
                    #print(file_sep_lst)
                    file_sep_lst[-1] = 'leftImg8bit.png'
                    
                    apath_lbl = os.path.join(maindir, filename)
                    
                    maindir_ori = ''
                    filename_ori = ''
                    for word in path_sep_lst:
                        maindir_ori = maindir_ori + word + '/'
                    
                    for word in file_sep_lst:
                        filename_ori = filename_ori + word +'_'
                    filename_ori = filename_ori[:-1]
                    
                    apath_ori = os.path.join(maindir_ori, filename_ori)
  
                    dst_file_lbl = 'test_' + '{0:05d}'.format(cnt_test) + '_lbl.png'
                    shutil.copy2(apath_lbl, obj_folder[2] + '/' + dst_file_lbl)
                    print('transfer name from ' + filename + ' to ' + dst_file_lbl)
                    
                    dst_file_ori = 'test_' + '{0:05d}'.format(cnt_test) + '_ori.png'
                    shutil.copy2(apath_ori, obj_folder[2] + '/' + dst_file_ori)
                    print('transfer name from ' + filename_ori + ' to ' + dst_file_ori)
                    
                    cnt_test += 1
                    
                    
                    
        if path_sep_lst[-2] == 'train':
            path_sep_lst[-3] = 'leftImg8bit'
          
            for filename in file_name_list:
                file_sep_lst = filename.split('_')
                
                if file_sep_lst[-1] == 'labelIds.png':
                    file_sep_lst.pop(-1)
                    #print(file_sep_lst)
                    file_sep_lst[-1] = 'leftImg8bit.png'
                    
                    apath_lbl = os.path.join(maindir, filename)
                    
                    maindir_ori = ''
                    filename_ori = ''
                    for word in path_sep_lst:
                        maindir_ori = maindir_ori + word + '/'
                    
                    for word in file_sep_lst:
                        filename_ori = filename_ori + word +'_'
                    filename_ori = filename_ori[:-1]
                    
                    apath_ori = os.path.join(maindir_ori, filename_ori)
  
                    dst_file_lbl = 'train_' + '{0:05d}'.format(cnt_train) + '_lbl.png'
                    shutil.copy2(apath_lbl, obj_folder[0] + '/' + dst_file_lbl)
                    print('transfer name from ' + filename + ' to ' + dst_file_lbl)
                    
                    dst_file_ori = 'train_' + '{0:05d}'.format(cnt_train) + '_ori.png'
                    shutil.copy2(apath_ori, obj_folder[0] + '/' + dst_file_ori)
                    print('transfer name from ' + filename_ori + ' to ' + dst_file_ori)
                    
                    cnt_train += 1
                    
                   
        if path_sep_lst[-2] == 'val':
            path_sep_lst[-3] = 'leftImg8bit'
          
            for filename in file_name_list:
                file_sep_lst = filename.split('_')
                
                if file_sep_lst[-1] == 'labelIds.png':
                    file_sep_lst.pop(-1)
                    #print(file_sep_lst)
                    file_sep_lst[-1] = 'leftImg8bit.png'
                    
                    apath_lbl = os.path.join(maindir, filename)
                    
                    maindir_ori = ''
                    filename_ori = ''
                    for word in path_sep_lst:
                        maindir_ori = maindir_ori + word + '/'
                    
                    for word in file_sep_lst:
                        filename_ori = filename_ori + word +'_'
                    filename_ori = filename_ori[:-1]
                    
                    apath_ori = os.path.join(maindir_ori, filename_ori)
  
                    dst_file_lbl = 'valid_' + '{0:05d}'.format(cnt_valid) + '_lbl.png'
                    shutil.copy2(apath_lbl, obj_folder[1] + '/' + dst_file_lbl)
                    print('transfer name from ' + filename + ' to ' + dst_file_lbl)
                    
                    dst_file_ori = 'valid_' + '{0:05d}'.format(cnt_valid) + '_ori.png'
                    shutil.copy2(apath_ori, obj_folder[1] + '/' + dst_file_ori)
                    print('transfer name from ' + filename_ori + ' to ' + dst_file_ori)
                    
                    cnt_valid += 1
                
file_transfer('/datasets/home/14/214/c6du/data/' + 'Cityscapes/gtFine')
```
